# 1. Merge Strings Alternately (LeetCode 1768)

Study plan: https://leetcode.com/studyplan/leetcode-75/

Problem: https://leetcode.com/problems/merge-strings-alternately/

## English
Given two strings, `word1` and `word2`, create a new string by taking characters from the two strings one by one in alternating order, starting with `word1`. If one string has extra characters after the other runs out, append those remaining characters to the end of the result.

Return the merged string.

## Русский
Даны две строки `word1` и `word2`. Нужно создать новую строку, добавляя символы из этих строк поочередно, начиная с `word1`. Если одна из строк закончится раньше, оставшиеся символы второй строки нужно дописать в конец результата.

Верните получившуюся строку.


```python
class Solution:
    def mergeAlternately(self, word1: str, word2: str) -> str:
        word_list = []
        min_len = min(len(word1), len(word2))
        tail_word1 = word1[min_len:]
        tail_word2 = word2[min_len:]

        for letter1, letter2 in zip(word1, word2):
            word_list.append(letter1)
            word_list.append(letter2)

        word_list.append(tail_word1 + tail_word2)

        return ''.join(word_list)
```

# 2. Greatest Common Divisor of Strings (LeetCode 1071)

Study plan: https://leetcode.com/studyplan/leetcode-75/

Problem: https://leetcode.com/problems/greatest-common-divisor-of-strings/

## English
We say a string `t` divides a string `s` if `s` can be built by concatenating `t` with itself one or more times.

Given two strings `str1` and `str2`, return the longest string `x` such that `x` divides both `str1` and `str2`.

## Русский
Говорим, что строка `t` «делит» строку `s`, если `s` можно получить как `t`, повторенную подряд один или более раз.

Даны строки `str1` и `str2`. Нужно вернуть самую длинную строку `x`, которая делит обе строки.


```python
from math import gcd

class Solution:
    def gcdOfStrings(self, str1: str, str2: str) -> str:
        if str1 + str2 != str2 + str1:
            return ""

        return str1[:gcd(len(str1), len(str2))]
```

# 3. Kids With the Greatest Number of Candies (LeetCode 1431)

Study plan: https://leetcode.com/studyplan/leetcode-75/

Problem: https://leetcode.com/problems/kids-with-the-greatest-number-of-candies/

## English
You are given an array `candies`, where `candies[i]` is the number of candies the i-th kid has, and an integer `extraCandies`.

Return a boolean array `result` where `result[i]` is `true` if giving all `extraCandies` to the i-th kid makes their total candies **at least** the maximum candies any kid has; otherwise `false`.

## Русский
Дан массив `candies`, где `candies[i]` — сколько конфет у i-го ребенка, и число `extraCandies` — сколько дополнительных конфет у нас есть.

Нужно вернуть булевый массив `result`: `result[i] = True`, если после того как мы отдадим i-му ребенку все `extraCandies`, у него будет конфет не меньше, чем максимум у любого другого ребенка; иначе `False`.


```python
from typing import List

class Solution:
    def kidsWithCandies(self, candies: List[int], extraCandies: int) -> List[bool]:
        max_candies = max(candies)
        return [candy + extraCandies >= max_candies for candy in candies]
```

# 4. Can Place Flowers (LeetCode 605)

Study plan: https://leetcode.com/studyplan/leetcode-75/

Problem: https://leetcode.com/problems/can-place-flowers/

## English
You have a flowerbed represented by an array `flowerbed` consisting of `0` and `1`, where `0` means empty and `1` means already planted.

You want to plant `n` new flowers, but **no two flowers can be in adjacent plots**.

Return `true` if it is possible to plant `n` flowers without breaking the rule; otherwise return `false`.

## Русский
Дана клумба в виде массива `flowerbed`, состоящего из `0` и `1`: `0` — место пустое, `1` — там уже растёт цветок.

Нужно посадить `n` новых цветков так, чтобы **ни два цветка не стояли рядом** (в соседних ячейках).

Верните `True`, если можно посадить `n` цветков, не нарушая это правило; иначе верните `False`.


```python
from typing import List

class Solution:
    def canPlaceFlowers(self, flowerbed: List[int], n: int) -> bool:
        count_flower = 0
        for i in range(len(flowerbed)):
            if flowerbed[i] == 0:
                left_ok = (i==0) or (flowerbed[i-1]==0)
                right_ok = (i==len(flowerbed)-1) or (flowerbed[i+1]==0)

                if left_ok and right_ok:
                    count_flower += 1
                    flowerbed[i] = 1
        return count_flower >= n
```

# 5. Reverse Vowels of a String

Study plan: https://leetcode.com/studyplan/leetcode-75/

Problem: https://leetcode.com/problems/reverse-vowels-of-a-string/

## English
You are given a string `s`. Modify it by reversing the order of all vowel characters, while leaving every non-vowel symbol in its original position. 

Vowels include both lowercase and uppercase forms of `'a', 'e', 'i', 'o', and 'u'`. The same vowel may appear multiple times in the input. Return the resulting string.

## Русский
Вам передаётся строка `s`. Измените её, развернув последовательность всех гласных символов, при этом все остальные знаки должны остаться на своих местах. 

К гласным относятся строчные и заглавные буквы: `'a', 'e', 'i', 'o', 'u'`. В исходной строке гласные могут повторяться. Верните полученную строку.

```python
class Solution:
    def reverseVowels(self, s: str) -> str:
        vowels = set("aeiouAEIOU")
        s_vowels = [vow for vow in s if vow in vowels]
        s_vowels.reverse()
        result_s = []

        iter_s_vowels = iter(s_vowels)
        for ind in range(len(s)):
            if s[ind] in vowels:
                result_s.append(next(iter_s_vowels))
                continue
            result_s.append(s[ind])
        return ''.join(result_s)
```

# 7. Product of Array Except Self (LeetCode 238)

Study plan: https://leetcode.com/studyplan/leetcode-75/

Problem: https://leetcode.com/problems/product-of-array-except-self/

## English
Given an integer array `nums`, return an array `answer` such that `answer[i]` is equal to the product of all the elements of `nums` except `nums[i]`.

The product of any prefix or suffix of `nums` is guaranteed to fit in a 32-bit integer.

You must write an algorithm that runs in `O(n)` time and without using the division operation.

## Русский
Дан массив целых чисел `nums`. Нужно вернуть массив `answer`, где `answer[i]` равен произведению всех элементов `nums`, кроме `nums[i]`.

Гарантируется, что произведение любого префикса или суффикса `nums` помещается в 32-битное целое.

Требуется алгоритм за `O(n)` без использования операции деления.

```python
class Solution:
    def productExceptSelf(self, nums: List[int]) -> List[int]:
        n = len(nums)
        answer = [1] * n
        
        left_acc = 1
        for i in range(n):
            answer[i] = left_acc
            left_acc *= nums[i]
        
        right_acc = 1
        for i in range(n - 1, -1, -1):
            answer[i] *= right_acc
            right_acc *= nums[i]
            
        return answer
```

# 8. Increasing Triplet Subsequence (LeetCode 334)

Study plan: https://leetcode.com/studyplan/leetcode-75/

Problem: https://leetcode.com/problems/increasing-triplet-subsequence/

## English
Given an integer array `nums`, return `true` if there exists a triple of indices `(i, j, k)` such that `i < j < k` and `nums[i] < nums[j] < nums[k]`. If no such indices exist, return `false`.

## Русский
Дан массив целых чисел `nums`. Верните `True`, если существуют индексы `(i, j, k)` такие, что `i < j < k` и `nums[i] < nums[j] < nums[k]`. Если таких индексов нет — верните `False`.

```python
class Solution:
    def increasingTriplet(self, nums: List[int]) -> bool:
        junior = float('inf')
        midle = float('inf')

        for num in nums:
            if num <= junior:
                junior = num
            elif num <= midle:
                midle = num
            else:
                return True
        
        return False
```

# 9. String Compression (LeetCode 443)

Study plan: https://leetcode.com/studyplan/leetcode-75/

Problem: https://leetcode.com/problems/string-compression/

## English
Given an array of characters `chars`, compress it using the following algorithm:

Begin with an empty string `s`. For each group of consecutive repeating characters in `chars`:

- If the group's length is `1`, append the character to `s`.
- Otherwise, append the character followed by the group's length.

The compressed string `s` should not be returned separately, but instead, be stored in the input character array `chars`. Note that group lengths that are `10` or longer will be split into multiple characters in `chars`.

After you are done modifying the input array, return the new length of the array.

You must write an algorithm that uses only constant extra space.

Note: The characters in the array beyond the returned length do not matter and should be ignored.

## Русский
Дан массив символов `chars`. Сожмите его по следующему правилу:

Начните с пустой строки `s`. Для каждой группы подряд идущих одинаковых символов в `chars`:

- Если длина группы равна `1`, добавьте символ в `s`.
- Иначе добавьте символ, а затем длину группы.

Строку `s` отдельно возвращать не нужно — результат должен быть записан прямо в исходный массив `chars`. Учтите, что числа `10` и больше записываются несколькими символами (например, `12` как `'1'`, `'2'`).

После модификации массива верните новую длину.

Требуется алгоритм с `O(1)` дополнительной памятью.

Примечание: элементы `chars` после возвращаемой длины не важны и должны игнорироваться.

```python

class Solution:
    def compress(self, chars: List[str]) -> int:
        save_letter = None
        count_letter = 1
        count_index = 0

        for letter in chars:
            if save_letter == None:
                save_letter = letter
                continue
            
            if save_letter == letter:
                count_letter += 1
            elif count_letter == 1:
                chars[count_index] = save_letter
                save_letter = letter
                count_index += 1
            else:
                chars[count_index] = save_letter
                count_index += 1

                for num in str(count_letter):
                    chars[count_index] = num
                    count_index += 1
                save_letter = letter
                count_letter = 1

        else:
            if count_letter == 1:
                chars[count_index] = save_letter
                count_index += 1
            else:
                chars[count_index] = save_letter
                count_index += 1
                for num in str(count_letter):
                    chars[count_index] = num
                    count_index += 1

        return count_index
```

# 10. Move Zeroes (LeetCode 283)

Study plan: https://leetcode.com/studyplan/leetcode-75/

Problem: https://leetcode.com/problems/move-zeroes/description/?envType=study-plan-v2&envId=leetcode-75

## English
Given an integer array `nums`, move all `0`'s to the end of it while maintaining the relative order of the non-zero elements.

Note that you must do this in-place without making a copy of the array.

## Русский
Дан массив целых чисел `nums`. Переместите все нули в конец массива, сохраняя относительный порядок ненулевых элементов.

Важно: модифицируйте массив на месте, не создавая копию.

```python
class Solution:
    def moveZeroes(self, nums: List[int]) -> None:
        """
        Do not return anything, modify nums in-place instead.
        """
        write = 0

        for ind, num in enumerate(nums):
            if num != 0:
                nums[write] = num
                write += 1

        for ind, num in enumerate(nums):
            if ind >= write:
                nums[ind] = 0
```

# 11. Is Subsequence (LeetCode 392)

Study plan: https://leetcode.com/studyplan/leetcode-75/

Problem: https://leetcode.com/problems/is-subsequence/description/?envType=study-plan-v2&envId=leetcode-75

## English
Given two strings s and t, return true if s is a subsequence of t, or false otherwise.

A subsequence of a string is a new string that is formed from the original string by deleting some (can be none) of the characters without disturbing the relative positions of the remaining characters. (i.e., "ace" is a subsequence of "abcde" while "aec" is not).

## Русский
Даны две строки s и t. Верните true, если s является подпоследовательностью t, иначе false.

Подпоследовательность строки — это новая строка, сформированная из исходной путём удаления некоторых (или ни одного) символов без нарушения относительного порядка оставшихся символов. (Например, "ace" является подпоследовательностью "abcde", а "aec" — нет).

```python
class Solution:
    def isSubsequence(self, s: str, t: str) -> bool:
        pointer_t = 0

        for letter_s in s:
            for ind_t, letter_t in enumerate(t[pointer_t:]):
                if letter_s == letter_t:
                    pointer_t += ind_t+1
                    break
            else: break
        else: return True

        return False
```

# 12. Container With Most Water (LeetCode 11)

Study plan: https://leetcode.com/studyplan/leetcode-75/

Problem: https://leetcode.com/problems/container-with-most-water/description/?envType=study-plan-v2&envId=leetcode-75

## English
You are given an integer array `height` of length `n`. There are `n` vertical lines drawn such that the two endpoints of the `i`th line are `(i, 0)` and `(i, height[i])`.

Find two lines that together with the x-axis form a container, such that the container contains the most water.

Return the maximum amount of water a container can store.

Notice that you may not slant the container.

## Русский
Дан массив целых чисел `height` длины `n`. Для каждого индекса `i` построена вертикальная линия с концами в точках `(i, 0)` и `(i, height[i])`.

Найдите две такие линии, которые вместе с осью x образуют контейнер с максимальным объёмом воды.

Верните максимальное количество воды, которое может вместить такой контейнер.

Обратите внимание: контейнеры нельзя наклонять.

```python
class Solution:
    def maxArea(self, height: List[int]) -> int:
        left_index = 0
        right_index = len(height) - 1
        max_area = 0

        while left_index < right_index:
            left_line = height[left_index]
            right_line = height[right_index]

            current_area = min(left_line, right_line) * (right_index - left_index)
            max_area = current_area if max_area < current_area else max_area

            if left_line < right_line:
                left_index += 1
            else:
                right_index -= 1
        
        return max_area
```

# 13. Max Number of K-Sum Pairs (LeetCode 1679)

Study plan: https://leetcode.com/studyplan/leetcode-75/

Problem: https://leetcode.com/problems/max-number-of-k-sum-pairs/

## English
You are given an integer array `nums` and an integer `k`.

In one operation, you can pick two numbers from the array whose sum equals `k` and remove them from the array.

Return the maximum number of operations you can perform on the array.

## Русский
Дан массив целых чисел `nums` и целое число `k`.

За одну операцию можно выбрать два числа из массива, сумма которых равна `k`, и удалить их из массива.

Верните максимальное количество операций, которое можно выполнить.

```python
class Solution:
    def maxOperations(self, nums: List[int], k: int) -> int:
        freq_nums = {}
        count_operations = 0

        for num in nums:
            target_num = k - num

            if freq_nums.get(target_num, 0) > 0:
                freq_nums[target_num] -= 1
                count_operations += 1
            else:
                freq_nums[num] = freq_nums.get(num, 0) + 1
        return count_operations
```

# 14. Maximum Average Subarray I (LeetCode 643)

Study plan: https://leetcode.com/studyplan/leetcode-75/

Problem: https://leetcode.com/problems/maximum-average-subarray-i/description/?envType=study-plan-v2&envId=leetcode-75

## English
You are given an integer array `nums` consisting of `n` elements, and an integer `k`.

Find a contiguous subarray whose length is equal to `k` that has the maximum average value and return this value. Any answer with a calculation error less than `10^-5` will be accepted.

## Русский
Дан целочисленный массив `nums`, состоящий из `n` элементов, и целое число `k`.

Найдите непрерывный подмассив длины `k`, который имеет максимальное среднее значение, и верните это значение. Допускается ошибка расчёта менее `10^-5`.

```python
class Solution:
    def findMaxAverage(self, nums: List[int], k: int) -> float:
        current_sum = sum(nums[:k])
        max_avg_subarray = current_sum

        for i in range(k-1, len(nums)):
            current_sum += nums[i] - nums[i-k]
            max_avg_subarray = max(current_sum, max_avg_subarray)

        return max_avg_subarray / k
```

# 15. Maximum Number of Vowels in a Substring of Given Length (LeetCode 1456)

Study plan: https://leetcode.com/studyplan/leetcode-75/

Problem: https://leetcode.com/problems/maximum-number-of-vowels-in-a-substring-of-given-length/description/?envType=study-plan-v2&envId=leetcode-75

## English
Given a string `s` and an integer `k`, return the maximum number of vowel letters in any substring of `s` with length `k`.

Vowel letters in English are `a`, `e`, `i`, `o`, and `u`.

## Русский
Дана строка `s` и целое число `k`. Верните максимальное количество гласных букв в любом подстроке `s` длины `k`.

Гласные буквы: `a`, `e`, `i`, `o`, `u`.

```python
class Solution:
    def maxVowels(self, s: str, k: int) -> int:
        vowels = set("aeiou")
        current = sum(1 for letter in s[:k] if letter in vowels)
        max_vowels = current
        
        for i in range(k, len(s)):
            if s[i - k] in vowels:
                current -= 1
            if s[i] in vowels:
                current += 1

            max_vowels = max(max_vowels, current)
        return max_vowels
```

# 16. Max Consecutive Ones III (LeetCode 1004)

Study plan: https://leetcode.com/studyplan/leetcode-75/

Problem: https://leetcode.com/problems/max-consecutive-ones-iii/description/?envType=study-plan-v2&envId=leetcode-75

## English
Given a binary array `nums` and an integer `k`, return the maximum number of consecutive `1`'s in the array if you can flip at most `k` `0`'s.

## Русский
Дан двоичный массив `nums` и целое число `k`. Верните максимальное количество подряд идущих единиц в массиве, если вы можете изменить на единицу не более `k` нулей.

```python
class Solution:
    def longestOnes(self, nums: List[int], k: int) -> int:
        max_lenght = 0
        count_zero = 0
        left_border = 0
    
        for right_border in range(len(nums)):
            if nums[right_border] == 0:
                count_zero += 1

            while count_zero > k:
                if nums[left_border] == 0:
                    count_zero -= 1
                left_border += 1

            max_lenght = max(max_lenght, (right_border - left_border) + 1)
        
        return max_lenght
```

# 17. Longest Subarray of 1's After Deleting One Element (LeetCode 1493)
Study plan: https://leetcode.com/studyplan/leetcode-75/

Problem: https://leetcode.com/problems/longest-subarray-of-1s-after-deleting-one-element/

## English
Given a binary array nums, you should delete one element from it.

Return the size of the longest non-empty subarray containing only 1's in the resulting array. Return 0 if there is no such subarray.

## Русский
Дан двоичный массив nums. Нужно удалить из него ровно один элемент.

Верните длину самой длинной непустой подстроки, состоящей только из 1. Если такой подстроки не существует, верните 0.

```python
class Solution:
    def longestSubarray(self, nums: List[int]) -> int:
        max_size_one = 0
        left_border = 0
        count_zero = 0

        for right_border in range(len(nums)):
            if nums[right_border] == 0:
                count_zero += 1

            while count_zero > 1:
                if nums[left_border] == 0:
                    count_zero -= 1
                left_border += 1

            current_len = right_border - left_border + 1
            max_size_one = max(max_size_one, current_len)

        return max_size_one - 1
```

# 18. Find the Highest Altitude (LeetCode 1732)

Study plan: https://leetcode.com/studyplan/leetcode-75/  

Problem: https://leetcode.com/problems/find-the-highest-altitude/

## English
There are `n + 1` points on a road trip with different altitudes. The biker starts at point `0` with altitude `0`.  
Given an integer array `gain` of length `n`, where `gain[i]` is the net gain in altitude between points `i` and `i + 1`, return the highest altitude of a point.

## Русский
Маршрут состоит из `n + 1` точек на разной высоте. Велосипедист начинает в точке `0` на высоте `0`.  
Дан массив целых чисел `gain` длины `n`, где `gain[i]` — чистый прирост высоты между точками `i` и `i + 1`. Верните наивысшую достигнутую высоту точки.

```python
class Solution:
    def largestAltitude(self, gain: List[int]) -> int:
        max_alt = 0
        current_alt = 0
        
        for delta in gain:
            current_alt += delta
            max_alt = max(max_alt, current_alt)
        return max_alt
```

# 19. Find Pivot Index (LeetCode 724)
Study plan: https://leetcode.com/studyplan/leetcode-75/

Problem: https://leetcode.com/problems/find-pivot-index/description/?envType=study-plan-v2&envId=leetcode-75

## English
Given an array of integers `nums`, calculate the pivot index of this array.

The pivot index is the index where the sum of all the numbers strictly to the left of the index is equal to the sum of all the numbers strictly to the right of the index.

If the index is on the left edge of the array, then the left sum is `0` because there are no elements to the left. This also applies to the right edge of the array.

Return the leftmost pivot index. If no such index exists, return `-1`.

## Русский
Дан массив целых чисел `nums`. Найдите в нём индекс опоры (pivot index).

Индекс опоры — это такой индекс, при котором сумма всех чисел строго слева от него равна сумме всех чисел строго справа от него.

Если индекс находится на левом краю массива, то сумма слева равна 0, потому что слева нет элементов. То же самое относится и к правому краю массива.

Верните самый левый индекс опоры. Если такого индекса нет, верните -1.

```python
class Solution:
    def pivotIndex(self, nums: List[int]) -> int:
        sum_nums = sum(nums)
        left_sum = 0

        for ind_num, num in enumerate(nums):
            if ind_num != 0:
                left_sum += nums[ind_num-1]

            right_sum = (sum_nums - num)/2
            if right_sum == left_sum:
                return ind_num
        else:
            return -1
```

## English
Given two 0-indexed integer arrays `nums1` and `nums2``, return a list answer of size 2 where:

`answer[0]` is a list of all distinct integers in `nums1` that are not present in `nums2`.
`answer[1]` is a list of all distinct integers in `nums2` that are not present in `nums1`.
Note that the integers in the lists may be returned in any order.

## Русский
Даны два массива целых чисел `nums1` и `nums2` с нулевой индексацией. Верните список answer длины 2, где:

`answer[0]` — список всех различных чисел из `nums1`, которых нет в `nums2`.
`answer[1]` — список всех различных чисел из `nums2`, которых нет в `nums1`.
Порядок элементов в списках может быть любым.

```python
class Solution:
    def findDifference(self, nums1: List[int], nums2: List[int]) -> List[List[int]]:
        set1 = set(nums1)
        set2 = set(nums2)

        result1 = list(set1 - set2)
        result2 = list(set2 - set1)

        return [result1, result2]
```

Problem: https://leetcode.com/problems/unique-number-of-occurrences/description/?envType=study-plan-v2&envId=leetcode-75

## English
Given an array of integers `arr`, return `true` if the number of occurrences of each value in the array is unique, or `false` otherwise.

## Русский
Дан массив целых чисел `arr`. Верните `true`, если количество вхождений каждого значения в массиве уникально (то есть никакие два разных значения не встречаются одинаковое число раз), и `false` в противном случае.

```python
class Solution:
    def uniqueOccurrences(self, arr: List[int]) -> bool:
        dict_nums = {}
        for num in arr:
            if dict_nums.get(num, 0):
                dict_nums[num] += 1
            else: dict_nums[num] = 1

        if len(set(dict_nums.values())) == len(dict_nums.values()):
            return True
        else:
            return False
```

# 22. Determine if Two Strings Are Close (LeetCode 1657)
Study plan: https://leetcode.com/studyplan/leetcode-75/

Problem: https://leetcode.com/problems/determine-if-two-strings-are-close/description/?envType=study-plan-v2&envId=leetcode-75

## English
Two strings are considered close if you can attain one from the other using the following operations:

- **Operation 1**: Swap any two existing characters.  
  For example, `abcde` → `aecdb`
- **Operation 2**: Transform every occurrence of one existing character into another existing character, and do the same with the other character.  
  For example, `aacabb` → `bbcbaa` (all `a`'s turn into `b`'s, and all `b`'s turn into `a`'s)

You can use the operations on either string as many times as necessary.

Given two strings, `word1` and `word2`, return `true` if `word1` and `word2` are close, and `false` otherwise.

## Русский
Две строки считаются «близкими», если можно получить одну из другой, используя следующие операции:

- **Операция 1**: Поменять местами любые два существующих символа.  
  Например, `abcde` → `aecdb`
- **Операция 2**: Заменить все вхождения одного существующего символа на другой существующий символ, и наоборот.  
  Например, `aacabb` → `bbcbaa` (все `a` становятся `b`, все `b` становятся `a`)

Операции можно применять к любой из строк сколько угодно раз.

Даны две строки, `word1` и `word2`. Верните `true`, если строки являются «близкими», и `false` в противном случае.

```python
from collections import Counter

class Solution:
    def closeStrings(self, word1: str, word2: str) -> bool:
        if len(word1) != len(word2):
            return False

        freq_letters1 = Counter(word1)
        freq_letters2 = Counter(word2)
        
        match_unique_letters = set(freq_letters1) == set(freq_letters2)
        freq_match = sorted(freq_letters1.values()) == sorted(freq_letters2.values())
        
        return match_unique_letters and freq_match
```

# 23. Equal Row and Column Pairs (LeetCode 2352)
Study plan: https://leetcode.com/studyplan/leetcode-75/

Problem: https://leetcode.com/problems/equal-row-and-column-pairs/description/?envType=study-plan-v2&envId=leetcode-75

## English
Given a **0-indexed** `n x n` integer matrix `grid`, return the number of pairs `(ri, cj)` such that row `ri` and column `cj` are equal.

A row and column pair is considered equal if they contain the same elements in the same order (i.e., an equal array).

## Русский
Дана целочисленная матрица `grid` размером `n x n` с **нулевой индексацией**. Верните количество пар `(ri, cj)`, таких что строка `ri` и столбец `cj` равны.

Пара строки и столбца считается равной, если они содержат одинаковые элементы в одном и том же порядке (то есть представляют собой одинаковые массивы).

```python
from collections import Counter

class Solution:
    def equalPairs(self, grid: List[List[int]]) -> int:
        row_counts = Counter(tuple(row) for row in grid)
        columns = list(zip(*grid))
        
        count = 0
        for col in columns:
            if col in row_counts:
                count += row_counts[col]
                
        return count
```

# 24. Removing Stars From a String (LeetCode 2390)
Study plan: https://leetcode.com/studyplan/leetcode-75/

Problem: https://leetcode.com/problems/removing-stars-from-a-string/description/?envType=study-plan-v2&envId=leetcode-75

## English
You are given a string `s`, which contains stars `*`.

In one operation, you can:
- Choose a star in `s`.
- Remove the closest non-star character to its left, as well as remove the star itself.

Return the string after all stars have been removed.

**Note:**
- The input will be generated such that the operation is always possible.
- It can be shown that the resulting string will always be unique.

## Русский
Дана строка `s`, которая содержит звёздочки `*`.

За одну операцию можно:
- Выбрать звёздочку в `s`.
- Удалить ближайший непробельный символ слева от неё, а также удалить саму звёздочку.

Верните строку после удаления всех звёздочек.

**Примечание:**
- Входные данные генерируются таким образом, что операция всегда возможна.
- Можно показать, что итоговая строка всегда будет уникальной.

```python
class Solution:
    def removeStars(self, s: str) -> str:
        stack = []

        for letter in s:
            if letter != '*':
                stack.append(letter)
            else:
                stack.pop()
        
        return ''.join(stack)
```

# 25. Asteroid Collision (LeetCode 735)
Study plan: https://leetcode.com/studyplan/leetcode-75/

Problem: https://leetcode.com/problems/asteroid-collision/description/?envType=study-plan-v2&envId=leetcode-75

## English
We are given an array `asteroids` of integers representing asteroids in a row. The indices of the asteroid in the array represent their relative position in space.

For each asteroid, the absolute value represents its size, and the sign represents its direction (positive meaning right, negative meaning left). Each asteroid moves at the same speed.

Find out the state of the asteroids after all collisions. If two asteroids meet, the smaller one will explode. If both are the same size, both will explode. Two asteroids moving in the same direction will never meet.

## Русский
Дан массив целых чисел `asteroids`, представляющий собой ряд астероидов. Индексы астероидов в массиве обозначают их относительное положение в пространстве.

Для каждого астероида абсолютное значение представляет его размер, а знак — направление движения (положительное означает движение вправо, отрицательное — влево). Каждый астероид движется с одинаковой скоростью.

Определите состояние астероидов после всех столкновений. Если два астероида встречаются, меньший взрывается. Если оба имеют одинаковый размер, взрываются оба. Два астероида, движущиеся в одном направлении, никогда не встретятся.

```python
class Solution:
    def asteroidCollision(self, asteroids: List[int]) -> List[int]:
        stack = []
        len_asteroids = len(asteroids)

        for ast_ind in range(len_asteroids):
            if asteroids[ast_ind] < 0 and stack:
                while stack:
                    if stack[-1] < 0:
                        stack.append(asteroids[ast_ind])
                        break
                    if abs(stack[-1]) < abs(asteroids[ast_ind]):
                        stack.pop()
                    elif stack[-1] == abs(asteroids[ast_ind]):
                        stack.pop()
                        break
                    else:
                        break
                else:
                    stack.append(asteroids[ast_ind])
            else:
                stack.append(asteroids[ast_ind])
        return stack
```

# 26. Decode String (LeetCode 394)
Study plan: https://leetcode.com/studyplan/leetcode-75/

Problem: https://leetcode.com/problems/decode-string/description/?envType=study-plan-v2&envId=leetcode-75

## English
Given an encoded string, return its decoded string.

The encoding rule is: `k[encoded_string]`, where the `encoded_string` inside the square brackets is being repeated exactly `k` times. Note that `k` is guaranteed to be a positive integer.

You may assume that the input string is always valid; there are no extra white spaces, square brackets are well-formed, etc. Furthermore, you may assume that the original data does not contain any digits and that digits are only for those repeat numbers, `k`. For example, there will not be input like `3a` or `2[4]`.

The test cases are generated so that the length of the output will never exceed `10^5`.

## Русский
Дана закодированная строка, верните её декодированную версию.

Правило кодирования: `k[encoded_string]`, где `encoded_string` внутри квадратных скобок повторяется ровно `k` раз. Обратите внимание, что `k` гарантированно является положительным целым числом.

Можно считать, что входная строка всегда корректна; в ней нет лишних пробелов, квадратные скобки расставлены правильно и т.д. Кроме того, можно считать, что исходные данные не содержат цифр, а цифры используются только для чисел повторения `k`. Например, не будет входных данных вроде `3a` или `2[4]`.

Тестовые случаи сгенерированы так, что длина результирующей строки никогда не превысит `10^5`.

```python
class Solution:
    def decodeString(self, s: str) -> str:
        stack = []
        current_str = ""
        current_num = 0
        
        for char in s:
            if char.isdigit():
                current_num = current_num * 10 + int(char)
                
            elif char == '[':
                stack.append((current_str, current_num))
                current_str = ""
                current_num = 0
                
            elif char == ']':
                prev_str, num = stack.pop()
                current_str = prev_str + current_str * num
                
            else:
                current_str += char
                
        return current_str
```

# 27. Number of Recent Calls (LeetCode 933)
Study plan: https://leetcode.com/studyplan/leetcode-75/

Problem: https://leetcode.com/problems/number-of-recent-calls/description/?envType=study-plan-v2&envId=leetcode-75

## English
You have a `RecentCounter` class which counts the number of recent requests within a certain time frame.

Implement the `RecentCounter` class:

- `RecentCounter()` Initializes the counter with zero recent requests.
- `int ping(int t)` Adds a new request at time `t`, where `t` represents some time in milliseconds, and returns the number of requests that has happened in the past 3000 milliseconds (including the new request). Specifically, return the number of requests that have happened in the inclusive range `[t - 3000, t]`.

It is guaranteed that every call to `ping` uses a strictly larger value of `t` than the previous call.

## Русский
У вас есть класс `RecentCounter`, который подсчитывает количество недавних запросов в определённом временном интервале.

Реализуйте класс `RecentCounter`:

- `RecentCounter()` Инициализирует счётчик нулём недавних запросов.
- `int ping(int t)` Добавляет новый запрос в момент времени `t`, где `t` — время в миллисекундах, и возвращает количество запросов, произошедших за последние 3000 миллисекунд (включая новый запрос). Конкретно, возвращает количество запросов, произошедших в диапазоне `[t - 3000, t]` включительно.

Гарантируется, что каждый вызов `ping` использует строго большее значение `t`, чем предыдущий вызов.

```python
class RecentCounter:

    def __init__(self):
        self.requests = []

    def ping(self, t: int) -> int:
        self.requests.append(t)

        while self.requests[0] < t - 3000:
            self.requests.pop(0)
        
        return len(self.requests)
```
